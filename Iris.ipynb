```python
# Import Librarys
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from pandas.plotting import parallel_coordinates
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# Load dataset
columns = ["spal length (cm)", "sepal width (cm)", "petal length (cm)" , "petal width", "Species"]
iris_df = pd.read_csv("iris_data.csv", names = columns)

sns.pairplot(iris_df, hue='Species')
```

```python
# Train a simple KNN classifier
iris = load_iris()
X = iris.data[:, :2]  # Using only two features for visualization
y = iris.target
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X, y)

# Plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, alpha=0.8, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', marker='o')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title("Decision Boundary for KNN (k=3)")
plt.show()
```

```python
# Load dataset
columns = ["spal length (cm)", "sepal width (cm)", "petal length (cm)" , "petal width", "Species"]
iris_df = pd.read_csv("iris_data.csv", names = columns)

X = iris_df.drop(columns=["Species"])
y = iris_df["Species"]

encoder = LabelEncoder()
y = encoder.fit_transform(y)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# Define model
tf_model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=(4,)),
    layers.Dropout(0.3),
    layers.Dense(32, activation='relu'),
    layers.Dropout(0.3),
    layers.Dense(3, activation='softmax')
])

# Compile with learning rate scheduling
optimizer = tf.keras.optimizers.Adam(learning_rate=0.005)
tf_model.compile(optimizer=optimizer, loss="sparse_categorical_crossentropy", metrics=["accuracy"])


# Train for 500 epochs with large batch size
tf_model.fit(X_train, y_train, epochs=500, batch_size=64, verbose=1)
```
